#### **Задача 1**

Ваша задача написать функцию `get_time_to_conversion`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает среднее время между заходом на сайт и целевым действием.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res`

**Решение**

Напишите свое решение ниже

```python
import json
from datetime import datetime
```

```python
df["session_"] = df["session"].apply(lambda x: json.loads(x))
```

```python
df["session_date"] = df["session_"].apply(lambda x: x[0])
```

```python
df["session_pages"] = df["session_"].apply(lambda x: x[2])
```

```python

def get_time_to_conversion(df):

    times = []

    for s in df["session"]:
        # распакуем куждую строку перевдя ее в формат списка
        start_str, duration, pages = json.loads(s)
        # переведем дату начала сессии в формат datatime
        start_date = datetime.fromisoformat(start_str)

        # список для хранения времени конверсии
        conv_times = []
        # отбираем время конверсии
        for page, ts in pages:
            if page == 'conversion':
                conv_times.append(datetime.fromisoformat(ts))
        # пропуск сессии
        if not conv_times:
            continue

        # ранее целевое действие
        conv_dt = min(conv_times)
        # разница в секундах между заходом и конверсией
        diff_sec = (conv_dt - start_date).total_seconds()
        times.append(diff_sec)

    res = round((sum(times)/len(times)),2)
    return res

res = get_time_to_conversion(df)
```

#### **Задача 3**

Напишите функцию `get_daily_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает cколько разных заходов в течение дня делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res3`

**Решение**

Напишите свое решение ниже

```python

def get_daily_session_count(df):
    df2 = df.copy()
    # извлекаем дату первой сессии
    df2['session_start'] = df2['session'].apply(lambda s: datetime.fromisoformat(json.loads(s)[0]))
    # переводим дату в формат date
    df2['session_date'] = df2['session_start'].dt.date
    # группируем по пользователю и дате сессии и считаем их количество
    session_per_day = df2.groupby(['user_id', 'session_date'])['session'].count()
    # возвращаем среднее значение по колчиеству сессий на пользователя за день
    return round(session_per_day.mean(), 2)

res3 = get_daily_session_count(df)
```

#### **Задача 4**

Напишите функцию `get_average_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает сколько заходов вообще в среднем делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res4`

**Решение**

Напишите свое решение ниже

```python

def get_average_session_count(df):
    session_count = df.groupby('user_id')['session'].count()
    return round(session_count.mean(),2)

res4 = get_average_session_count(df)
```
